# Logistic Regression Project 

In this project I will be working with a fake advertising data set, indicating whether or not a particular internet user clicked on an Advertisement. I will try to create a model that will predict whether or not they will click on an ad based off the features of that user.

This data set contains the following features:

* 'Daily Time Spent on Site': consumer time on site in minutes
* 'Age': cutomer age in years
* 'Area Income': Avg. Income of geographical area of consumer
* 'Daily Internet Usage': Avg. minutes a day consumer is on the internet
* 'Ad Topic Line': Headline of the advertisement
* 'City': City of consumer
* 'Male': Whether or not consumer was male
* 'Country': Country of consumer
* 'Timestamp': Time at which consumer clicked on Ad or closed window
* 'Clicked on Ad': 0 or 1 indicated clicking on Ad

# Logistic Regression

Now it's time to do a train test split, and train our model!

```python
# Splitting the data into training set and testing set using train_test_split
from sklearn.model_selection import train_test_split
```

```python
# Checking my dataset columns to see which columns I want to allocate to mt training data
ad_data.columns
```

```python
X = ad_data[['Daily Time Spent on Site', 'Age', 'Area Income','Daily Internet Usage', 'Male']]

y = ad_data['Clicked on Ad']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

**Now its time to train and fit a logistic regression model on the training set.**

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

## Predictions and Evaluations
**Now let's predict values for the testing data.**

```python
predictions = model.predict(X_test)
```

```python
# Creating a classification report for the model.
```

```python
from sklearn.metrics import classification_report
```

```python
print('Classification Report: ')
print('')
print(classification_report(y_test, predictions))
```

**As you can see the precision, recall, and the f1-score is pretty good for our model, and our logestic regression model has done a pretty good job on the test data predicting who has clicked on Ads and who has not!**
