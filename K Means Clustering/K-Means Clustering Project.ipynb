## K Means Cluster Creation:

Now it is time to create the Cluster labels!

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import sklearn

print("scikit-learn version:", sklearn.__version__)
```

```python
# Creating an instance of a K Means model with 2 clusters.
kmeans = KMeans(n_clusters=2, n_init=10)
```

```python
X = df.drop('Private', axis=1)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)
```

```python
# Fit the KMeans model
kmeans.fit(scaled_data)
```

** What are the cluster center vectors?**

```python
kmeans.cluster_centers_
```

## Evaluation

As I said at the beginning of this project, there is no perfect way to evaluate clustering if you don't have the labels, however since this is just an exercise, we do have the labels, so I take advantage of this to evaluate my clusters, keep in mind, you usually won't have this luxury in the real world.

```python
# Create a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.
def converter(cluster):
    if cluster=='Yes':
        return 1
    else:
        return 0
```

```python
df['Cluster'] = df['Private'].apply(converter)
```

```python
df.head()
```

**Now let's create a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels.**

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(df['Cluster'],kmeans.labels_))
print(classification_report(df['Cluster'],kmeans.labels_))
```

Not so bad considering the fact that k-means algorithm is purely using the features to cluster the universities into 2 distinct groups! Hopefully you can begin to see how K Means is useful for clustering un-labeled data!
